Програмний розв'язок:

```python
from sympy import symbols, diff

x = symbols('x')
f = (x**3)/3 + (x**2)/2 - 2*x
df = diff(f, x)

result_1 = df.subs(x, 1)
result_2 = df.subs(x, -1/2)

print(f"f'(1) = {result_1}")
print(f"f'(-1/2) = {result_2}")
```

Програмний розв'язок:

```python
from sympy import sqrt

x = symbols('x')
g = sqrt(x**2 + 1)
dg = diff(g, x)

result_1 = dg.subs(x, 1)
result_2 = dg.subs(x, -1/2)

print(f"g'(1) = {result_1}")
print(f"g'(-1/2) = {result_2}")
```

**Завдання 2**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -3*x**2 + 30*x

def f_prime(x):
    return -6*x + 30

max_point = 5
max_value = f(max_point)

x_vals = np.linspace(0, 10, 100)
y_vals = f(x_vals)
derivative_vals = f_prime(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='y = -3x^2 + 30x')
plt.plot(x_vals, derivative_vals, label="y' = -6x + 30", linestyle='dashed')
plt.scatter(max_point, max_value, color='red', marker='o', label=f'Максимум у x={max_point}')

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(max_point, color='gray', linestyle='--', linewidth=1, label='Точка максимуму')
plt.title('Функція та її похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
